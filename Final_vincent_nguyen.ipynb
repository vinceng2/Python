```python
#define your own Full Time Employee class here
#Full Time Employee class inherits from Employee class
#the order of the four required positional arguments should be: name,address,vehicle object,salary
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class FullTimeEmployee(Employee):
    
    # Initialize the constructor
    def __init__(self, name, address, vehicle, salary, *args):
        super().__init__(name, address, vehicle, *args)
        self.__salary = salary
    
    # Define get method for salary
    def get_salary(self):
        return self.__salary 
    
    # Define set methof for salary
    def set_salary(self, salary):
        self.__salary = salary 
      
    # Define compute compensation method
    def compute_compensation(self):
        annual_comp = self.__salary
        
        # If the salary is less than or equal to 45000, multiply salary by 18% and subtract it from the annual_comp
        if self.__salary <= 45000:
            annual_comp -= self.__salary * 0.18
        
        # Else if the salary is greater than 45000 but less than or equal to 82000
        elif self.__salary > 45000 and self.__salary <= 82000:
            
            # Multiply 45000 by 18 percent, add it to the difference of salary and 45000, and multiply it by 28%
            annual_comp -= 45000 * 0.18 + (self.__salary - 45000) * 0.28
        
        # Else, multiply the salary by 18%, add it to 37000 (difference between 82000 and 45000) and times it by 28%
        # Finally, add it to the difference between the salary and 82000 and multiply by 33%
        else:
            annual_comp -= 45000 * 0.18 + 37000 * 0.28 + (self.__salary - 82000) * 0.33
        
        # Divide the annual compensation by 52 weeks: weekly_comp
        weekly_comp = annual_comp / 52
        
        # Return the weekly compensation 
        return weekly_comp
    
    # Define compute reimbursement method
    def compute_reimbursement(self, annual_expense):
        
        # If the annual expense is no more than $10,000, set the total annual reimbursement to the annual expense 
        if annual_expense <= 10000:
            total_annual_reimbursement = annual_expense
        
        # Else, if the annual expense is more than $10,000, set the total annual reimbursement to $10,000 plus 50 percent
        # of the amount of the annual expense that exceeds $10,000
        else:
            total_annual_reimbursement = 10000 + (annual_expense - 10000) * 0.5
        
        # Divide the total annual reimbursement by 52 weeks in a year: total_weekly_reimbursement
        total_weekly_reimbursement = total_annual_reimbursement / 52
        
        # Return the total weekly reimbursement for the full time employee 
        return total_weekly_reimbursement
    
    # __str__ method 
    def __str__(self):
        data = ""
        data += f"\nDetails of this Full Time Employee are:"
        data += super().__str__()
        data += f"\nSalary: {self.__salary:.2f}"
        return data
```

```python
#define your own Hourly Employee class here
#Hourly Employee class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,hourly rate
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class HourlyEmployee(Employee):
    
    # Initialize the constructor
    def __init__(self, name, address, vehicle, hoursWorked, hourlyRate):
        super().__init__(name, address, vehicle)
        self.__hours_wrkd = hoursWorked
        self.__hourly_rate = hourlyRate
        
    # Define get methods for hours worked and hourly rate
    def get_hours_worked(self):
        return self.__hours_wrkd
    
    def get_hourly_rate(self):
        return self.__hourly_rate
    
    # Define set methods for hours worked and hourly rate
    def set_hours_worked(self, hoursWorked):
        self.__hours_wrkd = hoursWorked
        
    def set_hourly_rate(self, hourlyRate):
        self.__hourly_rate = hourlyRate

    # Define compute compensation method 
    def compute_compensation(self):
        
        # If the amount of hours worked is less than or equal to 40, multiply the amount of hours worked by the hourly rate
        if self.__hours_wrkd <= 40:
            return self.__hours_wrkd  * self.__hourly_rate
        
        # Else, for hours excess of 40 hours, the hourly rate is 1.8 times the regular hourly rate
        # The sum of the two is the weekly compensation
        else:
            return 40 * self.__hourly_rate + (self.__hours_wrkd  - 40) * self.__hourly_rate * 1.8
    
    # Define compute reimbursement method
    def compute_reimbursement(self, weekly_expense):
        if weekly_expense <= 100:
            total_weekly_reimbursement = weekly_expense
            
        else:
            total_weekly_reimbursement = 100
            
        return total_weekly_reimbursement
        
    # __str__ method
    def __str__(self):
        data = ""
        data += f"\nDetails of this Hourly Employee are:"
        data += super().__str__()
        data += f"\nHours Worked: {self.__hours_wrkd}; Hourly Rate: {self.__hourly_rate}"
        return data
```

```python
#define your own Consultant class here
#Consultant class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class Consultant(Employee):
    
    # Initialize the constructor
    def __init__(self, name, address, vehicle, hoursWorked, projectType, *args):
        super().__init__(name, address, vehicle, *args)
        self.__hours_wrkd = hoursWorked
        self.__project_type = projectType
        
    # Define get methods for hoursWorked and projectType
    def get_hours_worked(self):
            return self.__hours_wrkd
        
    def get_project_type(self):
            return self.__project_type
        
    # Define set methods for hoursWorked and projectType
    def set_hours_worked(self, hoursWorked):
            self.__hours_wrkd = hoursWorked
            
    def set_project_type(self, projectType):
            self.__project_type = projectType
        
    # Define compute compensation method 
    def compute_compensation(self):
        
        # If the project type is one, multiply the amount of hours worked by $55.00
        if self.__project_type == 1:
            return self.__hours_wrkd * 55.00
            
        # Else if the project type is two, multiply the amount of hours worked by $70.00
        elif self.__project_type == 2:
            return self.__hours_wrkd * 70.00
        
        # Else if the project type is three, multiple the amount of hours worked by $85.00
        elif self.__project_type == 3:
            return self.__hours_wrkd * 85.00
        
        # Else, print invalid project type 
        else:
            print("Invalid project type!")
            
    # Define compute reimbursement method
    def compute_reimbursement(self, weekly_expense):
        if self.__project_type == 1:
            total_weekly_reimbursement = weekly_expense * 1
            return total_weekly_reimbursement
    
        elif self.__project_type == 2:
            total_weekly_reimbursement = weekly_expense * 0.9
            return total_weekly_reimbursement
        
        elif self.__project_type == 3:
            total_weekly_reimbursement = weekly_expense * 0.8
            return total_weekly_reimbursement
            
        else:
            print("Invalid project type!")
    
    # __str__ method
    def __str__(self):
        data = ""
        data += f"\nDetails of this Consultant are:"
        data += super().__str__()
        data += f"\nHours Worked: {self.__hours_wrkd}; Project Type: {self.__project_type}"
        return data
```

```python
#define your own Management class here
#Management class inherits from both Full Time Employee class and Consultant class
#the order of the six required positional arguments should be: name, address,vehicle object, salary, hours worked, project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class Management(FullTimeEmployee, Consultant):
    
    # __init__ method
    def __init__(self, name, address, vehicle, salary, hoursWorked, projectType, *args):
        
        # super().__init__ method to initialize the instance variables from the FullTimeEmployee and Consultant classes
        super().__init__(name, address, vehicle, salary, hoursWorked, projectType, *args)
        
    # Define compute compensation method
    def compute_compensation(self):
        
        # Compensation from the full time employee: from_full_time
        from_full_time = FullTimeEmployee.compute_compensation(self)
        
        # Compensation from the consultant: from_consultant
        from_consultant = Consultant.compute_compensation(self)
        
        # Sum of the compensation from the full time employee and consultant: total_weekly_comp
        total_weekly_comp = from_full_time + from_consultant
        
        # Return the total weekly compensation 
        return total_weekly_comp

    # Define compute reimbursement method
    def compute_reimbursement(self, annual_expense, weekly_expense):
        
        # Annual expense from the full time employee: from_ft_emp_reimbursement
        from_ft_emp_reimbursement = FullTimeEmployee.compute_reimbursement(self, annual_expense)
        
        # Weekly expense from the consultant: from_consul_reimbursement 
        from_consul_reimbursement = Consultant.compute_reimbursement(self, weekly_expense)
        
        # Sum of the annual expense from the full time employee and weekly expense from the consultant: total_reimbursement
        total_reimbursement = from_ft_emp_reimbursement + from_consul_reimbursement
        
        # Return the total reimbursement
        return total_reimbursement

    # __str__ method
    def __str__(self):
        data = ""
        data += f"\nDetails of this Management are:"
        data += Employee.__str__(self)
        data += f"\nSalary: {FullTimeEmployee.get_salary(self):.2f}; Hours Worked: {Consultant.get_hours_worked(self)}; Project Type: {Consultant.get_project_type(self)}"     
        return data
```

```python
#complete this get_vehicle_input() function to prepare for the main menu application
#this function is used to ask for vehicle information and then return them: make,model,year,mileage
#this function doesn't take arguement(s)
#no need to handle exceptions of make input and model input
#handle exceptions to make sure year is a 4-digit positive number between 1900 and 2020
#handle exceptions to make sure mileage is a positive number
def get_vehicle_input():
    
    # Prompt the user to enter the vehicle make: veh_make
    veh_make = input('Enter the vehicle make: ')
    
    # Prompt the user to enter the vehicle model: veh_model
    veh_model = input('Enter the vehicle model: ')
    
    while (True):
        try:
            
            # Prompt the end-user to enter the vehicle year (four-digit positive integer): veh_year
            veh_year = int(input('Enter the vehicle year: '))
            
            # If the vehicle year is less than four characters, raise a value error
            if len(str(veh_year)) != 4:
                raise ValueError
            
            # If the vehicle year is not between 1900 and 2020, raise a value error
            elif (veh_year < 1900) or (veh_year > 2020):
                raise ValueError
            
            # If the vehicle year is less than zero (negative), raise a value error
            elif veh_year < 0:
                raise ValueError
            
            # Else, break out of the while loop
            else:
                break
        
        # Raise the value error and print the following statement
        except ValueError:
            print("Invalid integer. Please enter a positive four-digit integer in the format of 'yyyy' between 1900 and 2020.")
            continue
            
    while True:
        try:
            # Prompt the user to enter the vehicle mileage: veh_mileage    
            veh_mileage = float(input('Enter the vehicle mileage: '))
            
            # If the vehicle mileage is less than zero, raise a value error and break out of the while loop
            if veh_mileage <= 0:
                raise ValueError
            break
        
        # Raise a value error and print the following statement
        except ValueError:
            print("The vehicle mileage cannot be negative. Please enter a positive number.")
            raise ValueError
    
    # Return the vehicle make, model, year, and mileage
    return (veh_make, veh_model, veh_year, veh_mileage)
```

```python
#complete this read_file_data() function to prepare for the main menu application
#this function is used to read information from an existing database/binary file, which is empdata.dat in the final exam
#this function doesn't take argument(s)
#this function returns a list containing all employee objects stored in the database; each employee object is an element in the list; the list will also be used and updated later
#need to handle exceptions if file doesn't exist

# Import pickle
import pickle

# Define the read file data function: read_file_data
def read_file_data():
    """
    Read information from the empdata.dat database/binary file
    
        Parameters:
            None
            
        Returns:
             emp_obj_list (list): A list containing all employee objects stored in the database, empdata.dat
    """
    
    # Initialize an empty list: emp_obj_list
    emp_obj_list = []
    try:
        
        # Open the empdata.dat file as read-only: fb
        with open('empdata.dat', 'rb') as fb:
            
            # Start of the try block
            try: 
                
                # This while loop will be always entered since the conditional expression will always be True
                while (True): 
                    
                    # Append each employee object from the empdata.dat file to the employee object list 
                    emp_obj_list.append(pickle.load(fb))
                    
            # Except the EOF Error and pass it
            except EOFError:
                pass
            
    # If there is no file found, throw a file not found error and print it
    except FileNotFound: 
        print("File is not found!")
        
    # Return the employee object list 
    return emp_obj_list
```

```python
#complete option 1 related code here
#the idea is to walk through users to input required information
#the first and foremost step is to ask for emloyee type
#this function doesn't take arguement(s)
#this function prints the entered object and also returns the entered object
#some parts are provided so you don't need to change them (but you still can if you want to)

def run_option1():
    '''
    Prompt the end-user to enter the employee information
    
        Parameters:
            None
        
        Returns:
            an_emp (class object): The employee object entered by the end-user 
    
    '''
    #ask for employee type
    #exceptions need to be handled here to accept valid input 
    
    # Initialize a new list containing all the employee types: sel
    valid_emp_type_list = [1, 2, 3, 4]
    
    # Start of the while look where it is set to true 
    while (True):
        
        # Start of the try block
        try: 
            
            # Prompt the user to enter the type of the employee
            sel = int(input("Enter the type of the employee (1-Full Time; 2-Hourly; 3-Consultant; 4-Management): "))
            
            # If the user input is not in the valid employee type list, raise a value error
            if sel not in valid_emp_type_list:
                raise ValueError
                
            # Break out of the loop
            break
        
        # Throw the value error and prompt the user to select an option from one to four 
        except ValueError:
            print("Please select an option from 1, 2, 3, and 4.")
      
    # Call the previous get_emp_input() to ask for employee information and store the values      
    name, address = get_emp_input()
    
    # Call the previous get_vehicle_input() to ask for vehicle information and store the values
    vehiclemake, vehiclemodel, vehicleyear, vehiclemileage = get_vehicle_input()    
    
    # Create an object of Vehicle class using the prervious information
    a_vehicle = Vehicle(vehiclemake, vehiclemodel, vehicleyear, vehiclemileage)
    
    #depending on previous input, call the previous functions to ask for corresponding information
    #then use all the required information to create an object of a certain class 
    #type 1:annual salary
    #type 2:hours worked,hourly rate
    #type 3:hours worked, project type
    #type 4:annual salary, hours worked, project type
    if sel == 1:
        sal = get_full_time_input()
        an_emp = FullTimeEmployee(name, address, a_vehicle,sal)
    elif sel == 2:
        hw, hr = get_hourly_input()
        an_emp = HourlyEmployee(name, address, a_vehicle, hw, hr)
    elif sel == 3:
        hw, pt = get_consultant_input()
        an_emp = Consultant(name, address, a_vehicle, hw, pt)
    elif sel == 4:
        sal, hw, pt = get_management_input()
        an_emp = Management(name, address, a_vehicle, sal, hw, pt)
    
    print(an_emp)
    
    print("\n============================================================")        
    print("New employee entered successfully! Now going to the main menu.")
    print("==============================================================")     
    return an_emp
```

```python
#complete option 2 related code here
#the function is to print centain number of employees as required by users
#first choice:print all employees; second choice: print first 5 employees when there are at least 5 employees (print all if fewer than 5)
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#needs to handle exceptions when taking user seletions
def run_option2(emp_list):
    #type in your code
    '''
    Print certain number of employees as required by users
    
        Parameters:
            emp_list (list): The list created previously which is used to store employee objects
        
        Returns:
            None
    '''
    # Initialize the valid choice list: valid_choice_list
    valid_choice_list = [1, 2]
    
    # Start of the while loop where it is set to true 
    while (True):
        
        # Start of the try block
        try:
            
            # Prompt the user to see the information for all or the first five employees
            emp_choice_input = int(input("Do you want to see the information for all employees (input 1) or the first five employees (input 2)?: "))
            
            # If the user's input is not one or two, raise a ValueError and break out of the while loop 
            if emp_choice_input not in valid_choice_list:
                raise ValueError
            break
            
        # Throw a value error and ask the user to select one or two 
        except ValueError:
            print('Please select 1 or 2.')
    
    # The first choice is to print all employees if the user enters one
    if emp_choice_input == 1:
        print("=====================================================================")        
        print("Below is the information of all the employees stored in the database.")
        print("=====================================================================") 
        
        # Iterate through the employee list and print all the employees
        for emp in range(len(emp_list)):
            print(emp_list[emp])
            
    # Else if the user enters two, the second choice is to print the first five employees when there are at least five employees
    elif emp_choice_input == 2:
        print("===============================================================================")        
        print("Below is the information of the first five the employees stored in the database.")
        print("================================================================================") 
        
        # If there are at least five employees in the list, print the first five
        if len(emp_list) >= 5:
            
            # Iterate through the first five employees and print them 
            for emp in range(5):
                print(emp_list[emp])
    
        # Else, if there are less than five employees, print all of them 
        else:
            
            # Iterate through the entire employee list and print them 
            for emp in range(len(emp_list)):
                print(emp_list[emp])
                print("\n")
```

```python
#complete option 4 related code here
#the function is used to search for an employee by name and the searching is NOT case sensitive
#if there are some matching cases, print the information of all matched employees
#if no matchng cases, print "There is no employee matching the name you entered."
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function also returns a list of objects of the matched employees
#no need to handle exceptions
def run_option4(emp_list):
    '''
    Search for an employee by name where the search is not case sensitive
    
        Parameters:
            emp_list (list): The list created previously which is used to store employee objects
            
        Return:
            matched_emp_obj_list (list): A list of objects of the matched employees     
    '''
    # Initialize an empty list to contain the employees' names matched to what the user inputted: matched_name_list
    matched_name_list = []
    
    # Prompt the user to enter the name of the employee they want to search
    emp_search_input = input('Please enter the name of the employee that you want to search: ')
    
    # Iterate through each employee in the employee list 
    for i in emp_list:
        
        # If the searched employee matches one of the employee's name
        if emp_search_input.lower() in i.get_emp_name().lower():
            
            # Append the employee to the matched_name_list 
            matched_name_list.append(i)
    
    # If there are no matched employee in the matched name list, print the following statement
    if len(matched_name_list) == 0:
        print('There is no employee matching the name you entered.')
    
    # Else, print the information of all employees that match the name the user entered 
    else:
        print("==========================================================================")
        print("Below is the information of all employees that match the name you entered.")
        print("==========================================================================\n")
        
        # Iterate through each employee in the employee name list and print the employee
        for emp in matched_name_list:
            print(emp)
    
    # Return matched_name_list
    return matched_name_list
```

```python
#complete option 5 related code here
#the function is used to show some basic statistics. They are: 
#(1) the number of employees stored in the database
#(2) the highest weekly compensation
#(3) the mean weekly compensation
#(4) the number of employees who have a vehicle with over 100,000 mileage
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#no need to handle exceptions

def run_option5(emp_list):
    '''
    Display basic statistics
    
        Parameters:
            emp_list (list): The list created previously which is used to store employee objects
            
        Return:
            None
    '''
    # Initialize a compensation list
    compensation_list = []
    
    # Initialize a mileage list
    mileage_list = []
    
    # Iterate through each employee in the employee list
    for emp in emp_list:
        
        # Append each employee compensation to the compensation list
        compensation_list.append(emp.compute_compensation())
        
        # Append each employee's vehicle mileage to the mileage list
        mileage_list.append(emp.get_vehicle().get_mileage())
    
    # Find the max compensation from the compensation list: max_compensation
    max_compensation = max(compensation_list)
    
    # Find the mean compensation from the compensation list: mean_compensation
    mean_compensation = sum(compensation_list) / len(compensation_list)
    
    # Initialize 100000 as the mileage to benchmark against: x
    x = 100000
    
    # Count the number of vehicle mileages over 100,000: over_100k_mileage
    over_100k_mileage = len([i for i in mileage_list if i > x])
    
    #
    print("==========================================================")
    print("Below is the statistics of all employees in this database.")
    print("==========================================================")
    
    # Print the number of employees in the database
    print("There are {} employees stored in this employee database.".format(len(emp_list)))
    
    # Print the highest weekly compensation in the compensation list
    print("The highest weekly compensation is ${:0.2f}.".format(max_compensation))
    
    # Print the mean weekly compensation in the compensation list
    print("The mean weekly compensation is ${:0.2f}.".format(mean_compensation))
    
    # Print the number of employees who have vehicles with over 100,000 mileage
    print("The number of employees who have a vehicle with over 100,000 mileage is {}.".format(over_100k_mileage))
```

```python
#complete option 6 related code here
#the function is used to compute weekly reimbursement (assuming the user has the requirerd information)
#the first step is to ask for the name of the employee (you can call run_option4() here)
#the second step is to select one employee from the matched employees
#the third step is to ask for required information 
#this function doesn't return anything
#no need to handle exceptions
def run_option6(emp_list):
    '''
    Compute the weekly reimbursement. 
    First, ask for the name of the employee. 
    Second, select one employee from the matched employee list. 
    Third, ask for the required information.
    
        Parameters:
            emp_list (list): The list created previously which is used to store employee objects
            
        Return:
            None    
    
    '''
    # Frist, ask the user to enter the name of the employee from the employee list
    matched_list = run_option4(emp_list)
    
    # Second, prompt the user to select one employee to check 
    choice = int(input("Which employee do you want to check (enter a number)?"))
    
    # Third, ask for the required information, the reimbursement
    emp = matched_list[choice - 1]
    
    # If the employee object is an instance of the full time employee class
    if isinstance(emp, FullTimeEmployee):
        
        # Input the annual expense 
        annual_expense_input = float(input("What is your annual expense?: "))
        
        # Print the full time employee's weekly reimbursement 
        print('##########################################################')
        print('This employee should have a weekly reimbursement of ${:0.2f}'.format(emp.compute_reimbursement(annual_expense_input)))
    
    # Else if the employee object is an hourly employee class
    elif isinstance(emp, HourlyEmployee):
        
        # Prompt the user to enter the weekly expense
        weekly_expense_input = float(input("What is your weekly expense?: "))
        
        # Print the hourly employee's weekly expense 
        print('###########################################################')
        print('This employee should have a weekly reimbursement of ${:0.2f}.'.format(emp.compute_reimbursement(weekly_expense_input)))
    
    # Else if the employee object is a consultant
    elif isinstance(emp, Consultant):
        
        # Prompt the user to enter the weekly expense
        weekly_expense_input = float(input("What is your weekly expense?: "))
        
        # Print the consultant's weekly reimbursement 
        print('###########################################################')
        print('This employee should have a weekly reimbursement of ${:0.2f}.'.format(emp.compute_reimbursement(weekly_expense_input)))
    
    # Else if the employee is a manager 
    elif isinstance(emp, Management):
        
        # Prompt the user to enter the annual expense
        annual_expense_input = float(input("What is your annual expense?: "))
        
        # Prompt the user to enter the weekly expense
        weekly_expense_input = float(input("What is your weekly expense?: "))
        
        # Print the manager's weekly reimbursement 
        print('###########################################################')
        print('This employee should have a weekly reimbursement of ${:0.2f}.'.format(emp.compute_reimbursement(annual_expense_input, weekly_expense_input)))
```

```python
#complete the main application
#the application menu provides 7 options
#when user selects a valid option (except for 7), execute that option and come back to the menu again
#need to handle exceptions when taking user's choice
#this function doesn't take any argument
#it will work on the employ object list which is read from file empdata.dat
def run_menu_options():
    '''
    Provide seven options for the employee menu application. When the user selects a valid option (except for seven), execute it and come back to the menu again.
    
        Parameters:
            None
            
        Return:
            None
    '''
    #read employee objects in to a list
    emp_list = read_file_data()
    
    #display 7 options and ask users to select from them and execute the selection
    sel = 1
    while sel <=  7 and sel >= 1:
        #display options
        print("\n==== Menu ====")
        print("1. To add an employee")
        print("2. To print the name and address of employees")
        print("3. To print the employee name and compensation of all employees")
        print("4. To search for employees by name")
        print("5. To check the basic statistics of employees")
        print("6. To calculate the reimbursement of one employee")
        print("7. To exit program")
        
        #take selection input
        while (True):
            try:
                
                # Prompt the user to enter an option
                sel = int(input("Your selection is: "))
                
                # If the option is not within option one to seven, raise a value error and break out of the loop
                if sel not in range(1,8):
                    raise ValueError
                break
            
            # Throw the value error and print the following statement 
            except ValueError:
                print("\n=========================================")
                print("You must enter an integer betwee 1 and 7!")
                print("=========================================\n")
        
        #execute selection
        if sel == 1:
            emp_list.append(run_option1())
        elif sel == 2:
            run_option2(emp_list)
        elif sel == 3:
            run_option3(emp_list)
        elif sel == 4:
            run_option4(emp_list)
        elif sel == 5:
            run_option5(emp_list)
        elif sel == 6:
            run_option6(emp_list)
        elif sel == 7:  
            run_option7(emp_list)
        else:
            print("Invalid choice!")
```
