```python
'''
Q1. Write a Python program to add 'ing' at the end of a given string (length should be at least
3). If the given string already ends with 'ing' then add 'ly' instead. If the string length of
the given string is less than 3, leave it unchanged. You need to use an input() function to
take inputs. The testing samples come from separate cases so you don’t need to use a
loop to keep asking for inputs. 
'''

# Prompt the user to enter a string
user_input = input('Please enter a string: ')

# If the string ends with 'ing' then add 'ly' to the end of the string
if user_input.endswith('ing'):
    user_input += 'ly'

# If the string length is at least three and already ends with 'ing' then add 'ly' to the end of the string instead. If the string length is less than three, leave it unchanged
elif len(user_input) >= 3:
    user_input += 'ing'
    
# Print the new string
print(f"The new string is: {user_input}")
```

```python
'''
Q2. Write a Python function called has_sublist (), that accepts a list as input and checks if any
of its elements are themselves lists. You only need to check the following three testing
cases and print out the results as shown below. You can directly assign values to the
argument list and don’t need to use an input() function. 
'''

# Define the function has_sublist()
def has_sublist(lst):
    
    '''
    The function, has_sublist, takes the one argument, lst, and checks if any of its elements are lists themselves.
    
    '''
    
    # Set the result to False: result
    result = False
    
    # For x in the argument lst, if the type "x" is equaled to a list, set result to true and break out of the loop
    for x in lst:
        if(type(x) == list):
            result=True
            break
    
    # Print it is {result} that the list {lst} has a sublist in it
    print(f"It is {result} that the list {lst} has a sublist in it.")

# Test the three scenarios
has_sublist([1, 2, [3, 4]])
has_sublist(['a', 'b', 'c', 'd'])
has_sublist([1, 2, ('a', 'b')])
```

```python
'''
Q3. Write a Python function concat_list() that accepts a list containing string values and an
integer as input and prints out a new list where each element in the list is concatenated
with all integers from 1 to n. You only need to check the following testing case and print
out the results as shown below. You can directly assign values to the input list and integer
(i.e. no need to use an input() function). 
'''

# Define the function concat_list()
def concat_list(lst, n):
    
    '''
    The function, concat_list, accepts a list containing string values, lst, and an integer, n, as 
    input and prints out a new list where each element in the list is concatenanted with all integers from 1 to n.
    '''
    
    # Create a new list to store the output: new_list
    new_list = []
    
    # Create a nested for loop where for x in the lst
    for x in lst:
        
        # For i in the range starting from 1 to n + 1, append the string + the string value to the new list
        for i in range(1, n+1):
            new_list.append(x + str(i))
    
    # Print the input list is {lst}
    print(f"The input list is {lst}.")
    
    # Print the input integer is {n}
    print(f"The input integer is {n}.")
    
    # Print the new list is {new_list}
    print(f"The new list is {new_list}.")

# Test the scenario where the list is ['a', 'b', 'c'. 'd'] and the integer is 5
concat_list(['a', 'b', 'c', 'd'], 5)
```

```python
'''
Q4. Write a Python program that accepts a comma separated sequence of words as input and
prints the unique words in sorted form (alphanumerically).
'''

# Prompt the user to input a sequence of words separated by commas
user_input = input('Please enter a sequence of words separated by commas: ')

# If the user does have a comma in the input, print the sequence is not separated by commas
if ',' not in user_input:
    print('The sequence is not separated by commas')

# Else, split the user input using commas: comma_sep
# Sort it alphanumerically and discard any duplicate values in the list: sorted_user_input
# Print the new sorted list containing only the unique words in alphanumerically
else:
    comma_sep = user_input.split(',')
    sorted_user_input = sorted(set(comma_sep))
    print(f"The new sorted list is: {sorted_user_input}")
```

```python
'''
Q5. Step 1: Use list comprehension expression to create a list to include elements that satisfy
the following criteria: (1) All elements are integers between 1 and 30 (inclusive); (2) All
elements are not divisible by 2 nor by 5. Print out the list and move to the next step.

Step 2: Convert all elements that are divisible by 3 to their opposites (e.g. 3 to -3) and
print out the new list (step-2 list).

Step 3: Sum up the value of all positive elements in the step-2 list and print out the result.
You can use the following format to print out your results.
'''

# Step 1: Create a list expression for all integers between 1 and 30 that are not divisible by 2 or 5. Print out the list.
list1 = [a for a in range(1, 31) if a%2 != 0 and a%5 != 0]
print(f"My initial list is: {list1}")

# Step 2: Convert all integers divisible by 3 to their opposites (e.g., 3 becomes -3). Print the new list.
list2 = [-b if b % 3 == 0 else b for b in list1]
print(f"My step-2 list is: {list2}")

# Step 3: Sum all positive values in the second list and print out the result
list3 = sum([c for c in list2 if c > 0])
print(f"The sum of positive elements in my step-2 list is: {list3}.")
```
