# Vehicle Class

```python
# Initialize the vehicle class
class Vehicle:
    
    # Initialize the constructor
    def __init__(self, make, model, year, mileage):
        self.make = make
        self.model = model
        self.year = year
        self.mileage = mileage
    
    # Define get method for vehicle make 
    def get_make(self):
        return self.make
    
    # Define get method for vehicle model 
    def get_model(self):
        return self.model
    
    # Define get method for vehicle year
    def get_year(self):
        return self.year
    
    # Define get method for vehicle mileage 
    def get_mileage(self):
        return self.mileage
    
    # Define set method for vehicle make 
    def set_make(self, make):
        self.make = make
    
    # Define set method for vehicle model 
    def set_model(self, model):
        self.model = model
    
    # Define set method for vehicle year 
    def set_year(self, year):
        self.year = year
    
    # Define set method for vehicle mileage 
    def set_mileage(self, mileage):
        self.mileage = mileage
    
    # __str__ method
    def __str__(self):
        return "Make: {0}; Model: {1}; Year of Manufacture: {2}; Mileage: {3}".format(self.make, self.model, self.year, self.mileage)
```

# Employee Class

```python
# Initialize the employee class
class Employee:
    
    # Initiate a constructor
    def __init__(self, name, address, vehicle):
        self.emp_name = name
        self.emp_address = address
        
        # Vehicle data (vehicle object)
        self.vehicle = vehicle
    
    # Define get method for employee name
    def get_emp_name(self):
        return self.emp_name
    
    # Define get method for employee address
    def get_emp_addr(self):
        return self.emp_address
    
    # Define get method for vehicle 
    def get_vehicle(self):
        return self.vehicle
    
    # Define set method for employee name
    def set_emp_name(self, name):
        self.emp_name = name
    
    # Define set method for employee address
    def set_emp_addr(self, address):
        self.emp_address = address  
        
    # Define set method for vehicle
    def set_emp_vehicle(self, vehicle):
        self.vehicle = vehicle
        
    def __str__(self):
        return "Employee Name: {0}, Employee Address: {1}, Vehicle: {2}".format(self.emp_name, self.emp_address, self.vehicle)
```

# FullTimeEmployee Class

```python
# Initialize the full time employee class
class FullTimeEmployee(Employee):
    
    # Initialize the constructor
    def __init__(self, name, address, vehicle, salary):
        Employee.__init__(self, name, address, vehicle)
        self.salary = salary
    
    # Define get method for salary
    def get_salary(self):
        return self.salary 
    
    # Define set methof for salary
    def set_salary(self, salary):
        self.salary = salary 
      
    # Define get compensation method
    def get_compensation(self):
        annual_comp = self.salary
        
        # If the salary is less than or equal to 45000, multiply salary by 18% and subtract it from the annual_comp
        if self.salary <= 45000:
            annual_comp -= self.salary * 0.18
        
        # Else if the salary is greater than 45000 but less than or equal to 82000
        elif self.salary > 45000 and self.salary <= 82000:
            
            # Multiply 45000 by 18 percent, add it to the difference of salary and 45000, and multiply it by 28%
            annual_comp -= 45000 * 0.18 + (self.salary - 45000) * 0.28
        
        # Else, multiply the salary by 18%, add it to 37000 (difference between 82000 and 45000) and times it by 28%
        # Finally, add it to the difference between the salary and 82000 and multiply by 33%
        else:
            annual_comp -= 45000 * 0.18 + 37000 * 0.28 + (self.salary - 82000) * 0.33
        
        # Divide the annual compensation by 52 weeks: weekly_comp
        weekly_comp = annual_comp / 52
        
        # Return the weekly compensation 
        return weekly_comp
    
    # __str__ method 
    def __str__(self):
        data = ""
        data += f"Details of this Full Time Employee are:\n"
        data += f"Employee Name: {self.emp_name}; Employee Address: {self.emp_address} \n"
        data += f"Make: {self.vehicle.make}; Model: {self.vehicle.model}; Year of Manufacture: {self.vehicle.year}; Mileage: {self.vehicle.mileage} \n"
        data += f"Salary: {self.salary:.2f} \n"
        return data
```

# HourlyEmployee Class

```python
# Initialize the hourly employee class
class HourlyEmployee(Employee):
    
    # Initialize the constructor
    def __init__(self, name, address, vehicle, hoursWorked, hourlyRate):
        Employee.__init__(self, name, address, vehicle)
        self.hoursWorked = hoursWorked
        self.hourlyRate = hourlyRate
        
    # Define get methods for hours worked and hourly rate
    def get_hours_worked(self):
        return self.hoursWorked
    
    def get_hourly_rate(self):
        return self.hourlyRate
    
    # Define set methods for hours worked and hourly rate
    def set_hours_worked(self, hoursWorked):
        self.hoursWorked = hoursWorked
        
    def set_hourly_rate(self, hourlyRate):
        self.hourlyRate = hourlyRate

    # Define get compensation method 
    def get_compensation(self):
        
        # If the amount of hours worked is less than or equal to 40, multiply the amount of hours worked by the hourly rate
        if self.hoursWorked <= 40:
            return self.hoursWorked * self.hourlyRate
        
        # Else, for hours excess of 40 hours, the hourly rate is 1.8 times the regular hourly rate
        # The sum of the two is the weekly compensation
        else:
            return 40 * self.hourlyRate + (self.hoursWorked - 40) * self.hourlyRate * 1.8
        
    # __str__ method
    def __str__(self):
        data = ""
        data += f"Details of this Hourly Employee are:\n"
        data += f"Employee Name: {self.emp_name}; Employee Address: {self.emp_address} \n"
        data += f"Make: {self.vehicle.make}; Model: {self.vehicle.model}; Year of Manufacture: {self.vehicle.year}; Mileage: {self.vehicle.mileage} \n"
        data += f"Hours Worked: {self.hoursWorked}; Hourly Rate: {self.hourlyRate} \n"
        return data
```

# Consultant Class

```python
# Initialize the consultant class 
class Consultant(Employee):
    
    # Initialize the constructor
    def __init__(self, name, address, vehicle, hoursWorked, projectType):
        Employee.__init__(self, name, address, vehicle)
        self.hoursWorked = hoursWorked
        self.projectType = projectType
        
    # Define get methods for hoursWorked and projectType
    def get_hours_worked(self):
            return self.hoursWorked
        
    def get_project_type(self):
            return self.projectType
        
    # Define set methods for hoursWorked and projectType
    def set_hours_worked(self, hoursWorked):
            self.hoursWorked = hoursWorked
            
    def set_project_type(self, projectType):
            self.projectType = projectType
        
    # Define get compensation method 
    def get_compensation(self):
        
        # If the project type is one, multiply the amount of hours worked by $55.00
        if self.projectType == 1:
            return self.hoursWorked * 55.00
            
        # Else if the project type is two, multiply the amount of hours worked by $70.00
        elif self.projectType == 2:
            return self.hoursWorked * 70.00
        
        # Else if the project type is three, multiple the amount of hours worked by $85.00
        elif self.projectType == 3:
            return self.hoursWorked * 85.00
        
        # Else, print invalid project type 
        else:
            print("Invalid project type!")
             
    # __str__ method
    def __str__(self):
        data = ""
        data += f"Details of this Consultant are:\n"
        data += f"Employee Name: {self.emp_name}; Employee Address: {self.emp_address} \n"
        data += f"Make: {self.vehicle.make}; Model: {self.vehicle.model}; Year of Manufacture: {self.vehicle.year}; Mileage: {self.vehicle.mileage} \n"
        data += f"Hours Worked: {self.hoursWorked}; Project Type: {self.projectType} \n"     
        return data
```

# Who has/have the highest weekly compensation

```python
# Initialize a new list: employees_list
employees_list = [fte1, fte2, fte3, hrlye1, hrlye2, hrlye3, consul1, consul2, consul3]

# Initialize a new list: comp_list
comp_list = []

# Initialize a new list: emp_name_list
emp_name_list = []

# For each employee in the employee database
for x in employee_db:
    
    # Append each employee's weekly compensation to the comp_list
    comp_list.append(x.get_compensation())

# For each employee in the employee database
for y in employee_db:
    
    # If the max weekly compensation equals the employee's compensation
    if max(comp_list) == y.get_compensation():
          
        # Append his/her name to the employee name list
        emp_name_list.append(y.emp_name)

# Print the person who has the highest weekly compensation
print("The people who have the highest weekly compensation ${0:.2f} are(is): {1}".format(max(comp_list), emp_name_list))
```
