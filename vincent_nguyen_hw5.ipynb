# Q1

```python
%matplotlib notebook
%matplotlib notebook  

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Import math
import math

with open('carprices.txt') as carprices_txt:
    carprices_readline = carprices_txt.readlines()[1:]
    
    
    # Store the numbers (integers) read from the text file: carprices_list
    carprices_list = []
    
    # Iterate through each line of the text file
    for line in carprices_readline:
        try:
            
            # Convert each line to an integer, including the strings
            a = int(line)
            
            # If the string is not an integer
            if type(a) != int:
                
                # Raise a ValueError
                raise ValueError
            
            # Else, divide the integer by 10, round it to the nearest integer, and append it to the carprices_list
            else:
                carprices_list.append(math.floor(a/10))

        # If the string cannot be converted to an integer, raise a ValueError and print the following statement
        except ValueError:
            print("This data is invalid and will be ignored:", str(line))
            
        # Plot the histogram
        plt.hist(carprices_list, color='cornflowerblue', bins=50, range=(0,50), histtype='bar', ec='white')
        
        # X and Y axis labels
        plt.xlabel('Price in tens of thousands')
        plt.ylabel('Number of cars')
        
        # Histogram title
        plt.title("Frequency of Car Prices")
        
        # Show the plot
        plt.show()
```

# Q2

```python
%matplotlib notebook
%matplotlib notebook  

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Import numpy as np
import numpy as np

# Initialize empty lists for years, bears, dolphins, and whales
years = []
bears = []
dolphins = []
whales = []

# Read the text file into a Numpy array and set the type as a string
wildlife_data = np.loadtxt('Wildlife.txt', dtype=str)

# Delete the first row since it contains the heading 
wildlife_data_no_heading = np.delete(wildlife_data, (0,0), axis=0)

# Iterate through each column in the Numpy array
for col in wildlife_data_no_heading:
    
    # Append the first column, years, to the years list as an integer
    years.append(int(col[0]))
    
    # Append the second column, bears, to the bears list as an integer
    bears.append(int(col[1]))
    
    # Append the third column, dolphins, to the dolphins list as an integer
    dolphins.append(int(col[2]))
    
    # Append the last column, whales, to the whales list as an integer
    whales.append(int(col[3]))

# Set the figure size
# plt.figure(figsize=(3,3))

# Plot the bears data
plt.plot(years, bears, label='Bears')

# Plot the dolphins data
plt.plot(years, dolphins, label='Dolphins')

# Plot the whales data
plt.plot(years, whales, label='Whales')

# Set the title
plt.title('Wildlife Population Per Year')

# Set the X and Y axis labels
plt.xlabel('Years')
plt.ylabel('Wildlife Population')

# Display the legend on the upper left hand screen
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
%matplotlib notebook
%matplotlib notebook  

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Initialize the different car type list: cartype_list
cartype_list = ['sporty', 'van', 'compact', 'large', 'midsize', 'small']

# Initialize an empty list of the frequency of each car type: cartype_freq_list
cartype_freq_list = [0] * 6

# Open the cartype.txt file: c
c = open('cartype.txt')

# Skip the first row 
next(c)

# Read the new cartype.txt file without the header: raw_cartype_file
raw_cartype_file = c.read().split("\n")

# Iterate through each car type in the raw_cartype_file_split list and calculate the frequency of each car type
for c_type in raw_cartype_file:
    
    # If the car type is in the cartype_list
    if c_type in cartype_list:
        
        # Index the cartype_list and return the lowest index where the car type appears: a
        a = cartype_list.index(c_type)
        
        # Increase the said car type by one in the cartype_freq_list 
        cartype_freq_list[a] +=1

# Set the color for each label
colors = ['#06C2AC', 'blue', '#ED0DD9', '#06C2AC', 'blue', '#ED0DD9']

# Set the explode for each label
explode = (0.03, 0.03, 0.03, 0.03, 0.03, 0.03)

# Plot the pie chart
plt.pie(cartype_freq_list, labels = cartype_list, colors = colors, explode = explode, startangle = 90, radius = 1.2, autopct = '%2.2f%%')
        
# Display the plot
plt.show()
```

```python
%matplotlib notebook
%matplotlib notebook  

# Import matplotlib.pyplot as plt 
import matplotlib.pyplot as plt

# Put the x-axis data in a list: years
years = ['2003', '2004', '2005', '2006', '2007'] 

# Put line one data (y-axis) in a list: full_time_students 
full_time_students = [10000, 9500, 11100, 12000, 13300] 

# Put line two data (y-axis) in a list: part_time_students
part_time_students = [2000, 2100, 2050, 2700, 2550] 

# Plot two line charts: fig, (ax1, ax2)
fig, (ax1, ax2) = plt.subplots(2)

# Plot ax1
ax1.plot(years, full_time_students, label = 'Enrollment of full-time students')

# Set the title for ax1
ax1.set_title('Student Enrollment Per Year')

# Set the y-label for ax1
ax1.set_ylabel('Number of Students Enrolled')

# Set the legend for ax1 at the upper-right hand corner 
ax1.legend(loc='upper right')

# Plot ax2
ax2.plot(years, part_time_students, label = 'Enrollment of part-time students')

# Set the y-label for ax2
ax2.set_ylabel('Number of Students Enrolled')

# Set the legend for ax2 at the upper-right hand corner 
ax2.legend(loc='upper right')

# Display the plot 
plt.show()
```
